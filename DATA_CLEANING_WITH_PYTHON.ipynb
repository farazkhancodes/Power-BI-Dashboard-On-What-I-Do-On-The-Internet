```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
data=pd.read_json('DATA/BrowserHistory_Fixed.json',orient='records')
data.head()
```

```python
data.drop_duplicates(inplace=True)
data.head()
```

```python
data['date_time']= pd.to_datetime(data['time_usec'],unit='us',utc=True)
data.drop(['time_usec'],axis=1,inplace=True)
data['date']=data.date_time.dt.date
data.date=pd.to_datetime(data.date)
data['time']=data.date_time.dt.time
data.time=data.date_time.dt.time.astype(str).apply(lambda x:x[:8])
data.drop(['date_time'],axis=1,inplace=True)
data.head()
```

```python
#Removing Useless Data
print(data[data.title=='New tab'])
print(data[data.title=='New Tab'])
```

```python
to_drop=data[(data.title=='New tab')|(data.title=='New Tab')].index
data.drop(to_drop,inplace=True)
data
```

```python
data.columns
```

```python
#separting data cause it's easy to play with less number of columns
history=data[['page_transition','title','url','client_id','date','time']]
icons=data[['url','favicon_url']]
```

```python
with pd.ExcelWriter('DATA/Data.xlsx') as writer:
    history.to_excel(writer,sheet_name='Browser_History',index=False)
    icons.to_excel(writer,sheet_name='Icons',index=False)
```
